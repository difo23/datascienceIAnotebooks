# Adquirir datos de distintos archivos csv

```python
#data set para vino rojo
n= 332

def cargar_varios_csv( n):

    ceros  = "00"
    decena  = "0"
    centena = ""
        
    file = "distributed-data/"+ceros+""+"1"+".csv"
    main = "../python-ml-course/datasets"
    full= os.path.join( main , file)
    df_sensor_total = pd.read_csv(full)

    for i in range(2, n+1):
        
        if i >= 10 and i < 100 :
            ceros = decena
        if i >= 100 :
            ceros = centena
            
        file = "distributed-data/"+ceros+""+str(i)+".csv"
        full = os.path.join( main , file)
        df   = pd.read_csv(full)
        df_sensor_total = pd.concat([df_sensor_total, df], axis=0)
    return df_sensor_total
```

```python
cargar_varios_csv(n)
```
